## Prediction

```python
img_predict = cv2.imread("arac.jpg")
#gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
model2 = YOLO("runs/detect/train/weights/best.pt")
```

```python
results = model.predict(source=img_predict, conf=0.5) #Conf-Confident 0.5  (50%)'nin altında eminsen bunu tahmin olarak sayma.
```

```python
boxes = results[0].boxes
annotated_img = img_predict.copy()
```

```python
import matplotlib.pyplot as plt
```

```python
for box in boxes:
    x1, y1, x2, y2 = map(int, box.xyxy[0])

    h, w, _ = annotated_img.shape

    cv2.rectangle(annotated_img, (x1,y1), (x2,y2), color=(255,0,0), thickness=2)
```

```python
%matplotlib inline
plt.imshow(annotated_img)
plt.show()
```
