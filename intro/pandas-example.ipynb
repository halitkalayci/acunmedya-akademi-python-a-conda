## Veri Ön Hazırlığı

```python
df3.isnull().sum()
```

```python
df3.head(30)
```

```python
df3["Country"].mode()[0]
```

```python
#17,24
#df3 = pd.read_csv("Chocolate Sales2.csv")
df3["Country"] = df3["Country"].fillna(  df3["Country"].mode()[0]  )
```

```python
df3.isnull().sum()
```

```python
# 16.satıra bir alternatif syntax
df3["Boxes Shipped"].fillna(  df3["Boxes Shipped"].mean(), inplace=True  )
df3.isnull().sum()
```

```python
df3.head(5)
```

```python
##df3.to_csv("doldurulmus_csv.csv")
```

```python
df3["Amount"]
```

```python
# $5,320 -> 5320
df3["Amount"] = df3["Amount"].astype(str).str.replace("[$,]", "", regex=True).astype(float)
df3["Amount"]
```

```python
df3.isnull().sum()
```

```python
df3["Amount"] = df3["Amount"].fillna(  df3["Amount"].mean() )
df3.isnull().sum()
```

```python
#ForwardFill BackwardFill
df3["Date"] = df3["Date"].ffill().bfill()
df3.isnull().sum()
```
