```python
import pandas as pd
```

```python
df = pd.read_csv("HousingData.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df = df.fillna(df.mean()) ## Her alanı o alanın ortalaması ile doldur!
df.isnull().sum()
```

```python
df.head()
```

```python
X = df.drop(columns=["MEDV"])
y = df["MEDV"]
```

```python
!pip install scikit-learn
```

```python
X
```

```python
y
```

```python
# Veriyi train ve test split etmek.
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
# Random Seed -> Rastgeleliği belirleyen bir çıkış noktası gibi. random_state=42
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#Lineer Regresyon
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test) 
y_pred

# Git eğitildiğin gibi şu X_test verileriyle tahminler yap ve bana ver.
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# MAE 
mae = mean_absolute_error(y_test, y_pred)


# MSE 
mse = mean_squared_error(y_test, y_pred)

mae, mse
# Bu skorlar ne kadar düşükse modelim o kadar iyi çalışıyor!
```

```python
import matplotlib.pyplot as plot
```

```python
plot.scatter(y_test, y_pred, label="Tahminler", alpha=0.6)
plot.xlabel("Gerçek Ev Fiyatları")
plot.ylabel("Tahmin Edilen Fiyatlar")
plot.legend()

min_val = min(y_test.min(), y_pred.min())
max_val = max(y_test.max(), y_pred.max())

plot.plot([min_val,max_val], [min_val,max_val],color="red", linestyle="--")

plot.show()
```
